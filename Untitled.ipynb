```python
df = pd.read_csv('aapl.csv')
df.head(2)
```

## Data Preprocessing

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Daily_Return'] = df['Close'].pct_change()
```

```python
df['MA_10'] = df['Close'].rolling(window=10).mean()
df['MA_50'] = df['Close'].rolling(window=50).mean()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Visualize Distributions

```python
sns.distplot(df['High'])

plt.show()
```

```python
sns.set_theme(style="whitegrid")
```

```python
# Create the line plot
plt.plot(df['High'], label='High')
plt.plot(df['Low'], label='Low')

# Set the plot title and labels
plt.title('High and Low Prices')
plt.xlabel('Date')
plt.ylabel('Price')

# Show the legend
plt.legend()

# Display the plot
plt.show()
```

```python
heatmap_data = df.pivot("High", "Volume",)

# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='Blues', cbar=True)

# Set the plot title and labels
plt.title('Volume Heatmap')
plt.xlabel('Volume')
plt.ylabel('Date')

# Display the plot
plt.show()
```

```python
grouped_df = df.groupby('Date')['Daily_Return'].mean().reset_index()

# Separate positive and negative returns
positive_returns = grouped_df[grouped_df['Daily_Return'] > 0]
negative_returns = grouped_df[grouped_df['Daily_Return'] < 0]

# Create the grouped bar plot
plt.figure(figsize=(12, 8))
plt.bar(positive_returns['Date'], positive_returns['Daily_Return'], color='green', label='Positive Returns')
plt.bar(negative_returns['Date'], negative_returns['Daily_Return'], color='red', label='Nagative Returns')

# Set the plot title and labels
plt.title('Grouped Bar Plot of Daily Returns')
plt.xlabel('Date')
plt.ylabel('Daily Return')

# Ratate the x-axis tick labels for better readability 
plt.xticks(rotation=45)

# Show the legend
plt.legend()

# Display the plot
plt.show()
```
